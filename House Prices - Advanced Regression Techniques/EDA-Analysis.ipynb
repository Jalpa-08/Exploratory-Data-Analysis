```python
features_with_na=[features for features in df.columns if df[features].isnull().sum()>=1]
```

```python
##these are the features with nan value
features_with_na
```

```python
#df['Electrical'].isnull().mean()
for feature in features_with_na:
    print(feature,np.round(df[feature].isnull().mean()*100,4), '% missing values')
```

```python
plt.figure(dpi = 150)
sns.boxplot(df['SalePrice'])
```

```python
## let's plot some amazing diagrams
data=df.copy()
for feature in features_with_na:
    
    # let's make a variables that indicate 1 if the observation was missing or 0 otherwise
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    # Lets calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

# Numerical variables

```python
df.info()
```

```python
## list of numerical variables

df['SaleType'].dtypes!='O' # 'o' for object
```

```python
numerical_features=[feature for feature in df.columns if df[feature].dtypes!='O']
```

```python
print(len(numerical_features))
df[numerical_features].head()
```

```python
## Temporal Variables(Eg: Datetime Variables)
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature,df[feature].unique())
```

```python
plt.figure(dpi = 150)
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
### Here we will compare the difference between all years features with Salesprice
data=df.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]
        plt.figure(dpi = 150)
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

# Observations
## Within all the the yearwise information, it is clearly visible that the the New houses are more costlier that the older one. looks like a pareto or power law distribution\

# EDA Part 2

```python
## numerical variables
## Continuous variables and Discrete variables
decrete_features =[feature for feature in numerical_features if len(df[feature].unique()) <= 25]
print(len(decrete_features))
```

```python
df[decrete_features].head()
```

```python
data= df.copy()
for features in decrete_features:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
    
```

```python
continuous_features = [feature for feature in numerical_features if feature not in decrete_features+year_feature]
print(len(continuous_features))

```

```python
data=df.copy()
for feature in continuous_features:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
```

```python
## We will be using logarithmic transformation
data=df.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
## Outliers 
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
