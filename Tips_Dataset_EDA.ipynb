# Data Visualization

```python
#smoker.value_counts()
plt.figure(figsize=(10,6))
sns.countplot(x='smoker', data=df,palette='rainbow')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day', data=df,palette='RdBu_r')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data = df, x = 'day', hue = 'time',palette='RdBu_r')
```

```python
x = df['total_bill']
y = df['tip']
plt.scatter(x, y)
plt.xlabel('Total Bill', fontsize = 15)
plt.ylabel('Tip', fontsize = 15)
```

```python
plt.hist(df['tip'])
```

```python
df.day.value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df['total_bill'],palette='rainbow')
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(y = 'tip', x = 'time', data = df)
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(y = 'tip', x = 'time',hue='day',data = df)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(y = 'total_bill', x = 'day', data = df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.pairplot(df)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['total_bill'])
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True,cmap='YlGnBu')
```
