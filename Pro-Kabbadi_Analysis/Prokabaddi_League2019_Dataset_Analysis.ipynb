Q : How much Players of different position in PKL ?

```python
player_data[['Name','Position']]
```

```python
# Find total number of Player by Position

total_player = player_data.groupby(['Position'], as_index=False).count()[['Name','Position']]
total_player.columns=['Total Players','Position']


total_player
```

```python
# Plot for Total Players by their Position in PKL

fig, axx =plt.subplots(figsize=(12,6))
axx.set_title("Total Players by their Position in PKL")
sns.barplot(x='Total Players', y='Position',hue='Position', data=total_player, palette='Set1')
```

Q : Top Players of League ?

```python
# Position of the Players

pos=['All Rounder','Raider','Defender']
pos_play = player_data.loc[player_data['Position'].isin(pos)]
pos_play

# Find total number of Position

total_pos = pos_play.groupby(['Position'], as_index=False).count()[['Position','Name']]
total_pos.columns=['Position','Total Players']

# Find total position count of each type

top5pos = pos_play.groupby(['Position','Name'], as_index=False)[['Position']].count()
top5pos.columns=['Name', 'Position']

top5pos
```

```python
player_data['Position'].value_counts().sort_values(ascending=False).head(5)
player_data['Position'].value_counts().sort_values(ascending=False).head(5).plot(kind='pie',figsize=(10,6), 
title="Position of Players", autopct='%1.2f%%')
plt.axis('equal')
```

```python
player_data['Name'].value_counts().sort_values(ascending=False).head(5)
player_data['Name'].value_counts().sort_values(ascending=False).head(5).plot(kind='pie',figsize=(10,6), 
title="Top 5 Players", autopct='%1.2f%%')
plt.axis('equal')
```
